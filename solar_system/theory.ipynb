### Закон всемирного тяготения

Сила $F$ гравитационного притяжения между двумя материальными точками с массами $m_1$ и $m_2$, разделёнными расстоянием $r$, действует вдоль соединяющей их прямой, пропорциональна обеим массам и обратно пропорциональна квадрату расстояния.

$$F = G·\cfrac{m_1 · m_2}{r^2}$$

Здесь $G$ - гравитационная постоянная, численно равная $6,67430(15)·10^{−11}\cfrac{м^3}{кг·с^2}$

### Второй закон Ньютона

В инерциальных системах отсчёта ускорение $\overrightarrow{a}$, приобретаемое материальной точкой массой $m$, прямо пропорционально вызывающей его силе $\overrightarrow{F}$, совпадает с ней по направлению и обратно пропорционально массе материальной точки.

$$\overrightarrow{a} = \cfrac{\overrightarrow{F}}{m}$$

## Движение планет в солнечной системы

Пользуясь двумя вышеприведенными формулами можно с достаточной точносью рассчитать движение планет солнечной системы вокруг нашей звезды. Для этого достаточно составить дифференциальное уравнение первого порядка (приведено ниже), которое впоследствии решается методом Рунге-Кутты четвертого порядка.

### 1. Положение тела в пространстве

Для описания положения космических тел мы будем считать, что они движутся в одной плоскости, пренебрегая некоторым наклоном орбит. Таким образом, нам достаточно двух координат $x$ и $y$ для того, чтобы задать местопложение планеты или любого другого космического тела в солнечной системе. Помимо этого, нам необходимо будет знать скорости наших объектов $V_x$ и $V_y$.

### 2. Система уравнений

Законы Ньютона (сначала второй, а потом первый), примененные для $n$ тел, дают нам следующую систему из $2n$ дифференциальных уравнений второго порядка (в общем виде):

$$
\begin{equation}
\begin{cases}
x_i'' = f_{ix}(x_1, ..., x_n, x_1', ..., x_n', t) \\
\\
y_i'' = f_{iy}(y_1, ..., y_n, y_1', ..., y_n', t)
\end{cases}
\end{equation}
$$

где $i = \overline{1, n}$, $x_i$ - положение тела $i$, $x_i'$ - его скорость, $x_i''$ - ускорение

* На самом деле, уравнения не зависит от времени, но для общего вида записал

### 3. Явный вид функций $f_i$

$$f_i = \frac{1}{m_i}\left(G·\frac{m_1·m_i}{R_{1i}^2} +\ \ ...\ \ +\ \ G·\frac{m_{i-1}·m_i}{R_{(i-1)i}^2}\ \ +\ \ G·\frac{m_{i+1}·m_i}{R_{(i+1)i}^2}\ \ +\ \ ...\ \ + G·\frac{m_n·m_i}{R_{ni}^2}\right)$$

Немного упростив получаем:

$$f_i = G\left(\frac{m_1}{R_{1i}^2}\ \ +\ \ ...\ \ +\ \ \frac{m_{i-1}}{R_{(i-1)i}^2}\ \ +\ \ \frac{m_{i+1}}{R_{(i+1)i}^2}\ \ +\ \ ...\ \ +\ \ \frac{m_n}{R_{ni}^2}\right)$$

Тогда в проекция на ось $x$ и $y$ имеем:

$$f_{ix} = G\left(\frac{m_1}{R_{1i}^2}\cos{\alpha_1}\ \ +\ \ ...\ \ +\ \ \frac{m_{i-1}}{R_{(i-1)i}^2}\cos{\alpha_{i-1}}\ \ +\ \ \frac{m_{i+1}}{R_{(i+1)i}^2}\cos{\alpha_{i+1}}\ \ +\ \ ...\ \ +\ \ \frac{m_n}{R_{ni}^2}\cos{\alpha_n}\right)$$

где $\cos{\alpha_j} = \cfrac{x_j - x_i}{R_{ji}}$

* Аналогчиная формула для f_{iy}, только вместо косинусов используются синусы

И финальный вид:

$$f_{ix} = G\left(\frac{m_1·(x_1-x_i)}{R_{1i}^3}\ \ +\ \ ...\ \ +\ \ \frac{m_{i-1}·(x_{i-1} - x_i)}{R_{(i-1)i}^3}\ \ +\ \ \frac{m_{i+1}·(x_{i+1} - x_i)}{R_{(i+1)i}^3}\ \ +\ \ ...\ \ +\ \ \frac{m_n·(x_n - x_i)}{R_{ni}^3}\right)$$

$$f_{iy} = G\left(\frac{m_1·(y_1-y_i)}{R_{1i}^3}\ \ +\ \ ...\ \ +\ \ \frac{m_{i-1}·(y_{i-1} - y_i)}{R_{(i-1)i}^3}\ \ +\ \ \frac{m_{i+1}·(y_{i+1} - y_i)}{R_{(i+1)i}^3}\ \ +\ \ ...\ \ +\ \ \frac{m_n·(y_n - y_i)}{R_{ni}^3}\right)$$

### 4. Немного преобразований

Учтем связь положения тела, его скорости, ускорения и получим, что уравнение $x_i'' = f_{ix}$ эквивалентно следующей системе из 2-ух дифференциальных уравнений 1-ого порядка:

$$
\begin{equation}
\begin{cases}
x_i' = V_{ix}\\
V_{ix}' = f_{ix}
\end{cases}
\end{equation}
$$

### 5. Финальная система уравнений

Таким образом, мы получаем финальную систему из $6n$ дифференциальных уравнений 1-ого порядка:

$$
\begin{equation}
\begin{cases}
x_1' = V_{1x}\\
y_1' = V_{1y}\\
V_{1y}' = G\left(\cfrac{m_2·(x_2-x_1)}{R_{21}^3}\ \ +\ \ ...\ \ +\ \ \cfrac{m_n·(x_n - x_1)}{R_{n1}^3}\right)\\
V_{1y}' = G\left(\cfrac{m_2·(y_2-y_1)}{R_{21}^3}\ \ +\ \ ...\ \ +\ \ \cfrac{m_n·(y_n - y_1)}{R_{n1}^3}\right)\\
...\\
x_n' = V_{nx}\\
y_n' = V_{ny}\\
V_{ny}' = G\left(\cfrac{m_1·(x_1-x_n)}{R_{1n}^3}\ \ +\ \ ...\ \ +\ \ \cfrac{m_{n-1}·(x_{n-1} - x_n)}{R_{(n-1)n}^3}\right)\\
V_{ny}' = G\left(\cfrac{m_1·(y_1-y_n)}{R_{1n}^3}\ \ +\ \ ...\ \ +\ \ \cfrac{m_{n-1}·(y_{n-1} - y_n)}{R_{(n-1)n}^3}\right)\\
\end{cases}
\end{equation}
$$

### 6. Векторная форма

Положим, что $$U = \begin{bmatrix} x_1 \\ y_1 \\ V_{1x} \\ V_{1y} \\ ... \\ x_n \\ y_n \\ V_{nx} \\ V_{ny} \end{bmatrix}$$

Тогда имеем:

$$U'= F(U)$$

Важные замечания:

* Для решения данной системы мы можем воспользоваться методом Рунге-Кутты четвертого порядка точности
* В уравнениях нигде не фигурирует ускорение, все что нам нужно: координаты объектов и их скорости
